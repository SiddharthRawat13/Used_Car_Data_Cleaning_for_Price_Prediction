```python
import pandas as pd
```

```python
df1 =df
```

```python
#seprating Engine column in three column 'engine, cylinder, Enginetype'

# Extract engine size (e.g., 6.2L, 1.6L)
df1['engine size'] = df1.Engine.str.extract(r'(\d+\.\d+L)')

# Extract number of cylinders (e.g., 8, 4)
df1['cylinder'] = df1['Engine'].str.extract(r'(\d+)-Cylinder')

# Extract engine type (e.g., Gas Engine, Turbo Gas Engine)
df1['engine type'] = df1['Engine'].str.split().str[-2:].apply(lambda x: ' '.join(x) if isinstance(x,list) else '')
```

```python
df1 = df1.drop('Engine',axis=1)
```

```python
#sepration Average column in two columns city milage and highway milage

df1[['city mileage','highway mileage']] = df1.Average.str.split('/', expand=True)
df1['city mileage'] = df1['city mileage'].str.replace('City', '')
df1['highway mileage'] = df1['highway mileage'].str.replace('Highway', '')
df1['engine type'] = df1['engine type'].str.replace('Engine', '')

```

```python
df1 = df1.drop('Average', axis=1)
```

```python
#sepration Transmission column in two column Gear, transmission Type
df1['gear'] = df1['Transmission'].str.extract(r'(\d+)-Speed', expand=False)
df1['transmission type'] = df1['Transmission'].str.extract(r'(Automatic|Manual)', expand=False)
```

```python
df1 = df1.drop('Transmission',axis=1)
```

```python
# change data tyme of average column and seprating it with city and highway column
#df1['city mileage'] = pd.to_numeric(df1['city mileage'], errors='coerce')

#df['city mileage']=df['city mileage'].replace('City','')
#df['highway mileage']=df['highway mileage'].replace('Highway','')
```

```python
df1['Miles'] = df1['Miles'].str.replace('miles', '').str.replace(',','')
```

```python
#removing ',' in price and change data type
df1['Price'] = pd.to_numeric(df1['Price'].str.replace(',', ''), errors='coerce')
```

```python
df1['engine size'] = pd.to_numeric(df1['engine size'], errors='coerce')
```

```python
#convert datatypes

numeric_columns  = ['cylinder', 'city mileage', 'highway mileage','gear']
df1[numeric_columns] = df1[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
#filling mean in null values

df1['engine size'].fillna(df1['engine size'].mean(),inplace=True)
df1['cylinder'].fillna(df1['cylinder'].mean(),inplace=True)
df1['city mileage'].fillna(df1['city mileage'].mean(),inplace=True)
df1['highway mileage'].fillna(df1['highway mileage'].mean(),inplace=True)
```

```python
#filling  mode in null values
df1['gear'].fillna(df1['gear'].mode()[0],inplace=True) 
df1['transmission type'].fillna(df1['transmission type'].mode()[0],inplace=True)
```

```python
df1['Name'] = df1['Name'].str.replace('Certified','')
```

```python
# Assuming df is your DataFrame
df1['engine size'] = df1['engine size'].str.rstrip('L')  # Remove 'L' suffix
df1['engine size'] = pd.to_numeric(df1['engine size'], errors='coerce')  # Convert to numeric

# If you prefer to handle the 'L' suffix in one step:
# df['engine size'] = pd.to_numeric(df['engine size'].str.rstrip('L'), errors='coerce')
```

```python
# Round values in all specified columns to the nearest integer
df1[['engine size', 'cylinder', 'city mileage', 'highway mileage', 'gear']] = df1[['engine size', 'cylinder', 'city mileage', 'highway mileage', 'gear']].round()

# Display DataFrame with rounded values
print(df1)
```
